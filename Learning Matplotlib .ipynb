```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize =(8,5))
plt.title("Gas prices(1990-2007)")

plt.plot(gas['Year'],gas['USA'],'b--')
plt.plot(gas['Year'],gas['Italy'],'r.-')
plt.plot(gas['Year'],gas['UK'])

#for i in gas:
#    if(i!='Year'):
#        plt.plot(gas.Year,gas[i])                                      

plt.xticks(gas['Year'][::3])                                            #indexing in columns
plt.xlabel("Years")
plt.legend(["USA","Itlay","UK"])
plt.show()
```
